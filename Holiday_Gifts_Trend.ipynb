```python
# Importing necessary libraries
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize=(19.2, 10.8))  # Adjust for 100 DPI
#Scraped Data from Sources (https://www.mytotalretail.com/article/holiday-shopping-trends-2024-inflation-experiences-and-the-consumers-new-priorities/)
# Adding more years of data for comparison
categories = ["Experiences", "Material Gifts", "Food & Beverages"]
years = ["2021", "2022", "2023", "2024"]

# Spending percentages for each category by year
spending_data = {
    "2021": [35, 55, 10],
    "2022": [37, 53, 10],
    "2023": [40, 50, 10],
    "2024": [48, 45, 7]
}

# X-axis positions for categories
x = np.arange(len(categories))
width = 0.2  # Width of each bar

# Plotting the data for multiple years
fig, ax = plt.subplots(figsize=(12, 6))

# Creating bars for each year
bars = []
for i, year in enumerate(years):
    offset = (i - len(years) / 2) * width + width / 2  # Adjusting position for each year
    bars.append(ax.bar(x + offset, spending_data[year], width, label=year))

# Adding labels, title, and custom ticks
ax.set_xlabel("Spending Categories", fontsize=12)
ax.set_ylabel("Percentage of Spending (%)", fontsize=12)
ax.set_title("Holiday Spending Trends by Year: 2021-2024", fontsize=16, weight='bold')
ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=12)
ax.legend(title="Year", fontsize=10)

# Adding value annotations on the bars
def add_annotations(bars_group, years):
    for i, bar_group in enumerate(bars_group):
        for bar in bar_group:
            height = bar.get_height()
            ax.annotate(f'{height}%',
                        xy=(bar.get_x() + bar.get_width() / 2, height),
                        xytext=(0, 3),  # Offset for text
                        textcoords="offset points",
                        ha='center', va='bottom', fontsize=9)

add_annotations(bars, years)

# Show the plot

#plt.tight_layout()
plt.show()

```
